## 多板塊

```python
import pandas as pd
from ivan_ptt import crawl_ptt_page_02

boards = ["womentalk", "Gossiping"]
df_all = pd.DataFrame()  # 先準備一個空的容器
for i in boards:
    df = crawl_ptt_page_02(Board_Name=i, page_num=1)
    df_all = pd.concat([df_all, df], axis=0)  # 把新的結果存進容器

# 存檔
df_all.to_csv(
    "ptt資料.csv",  # 檔案名稱
    encoding="utf-8",  # 編碼
    index=False,  # 是否保留index
)
```
