### 萃取關鍵字


```python
import jieba

jieba.set_dictionary("dict/dict.txt.big")
jieba.load_userdict("dict/user_dict.txt")
```

```python
import jieba.analyse

ptt_data = ptt_data.dropna(subset=["所有文"])
ptt_data["重要關鍵字"] = ptt_data["所有文"].apply(
    lambda x: jieba.analyse.extract_tags(str(x), topK=5)
)

ptt_data
```

```python
with open("dict/stopwords.txt", "r", encoding="utf-8-sig") as f:
    stops = set(f.readlines())


ptt_data["重要關鍵字stopwords"] = ptt_data["重要關鍵字"].apply(
    lambda x: [i for i in x if i not in stops]
)

ptt_data
```

### Apriori 演算法


```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
# 將重要關鍵字放入做 Apriori 關聯分析
records = list(ptt_data["重要關鍵字stopwords"])

records[:5]
```

```python
# TransactionEncoder進行one_hot編碼。
import numpy as np

te = TransactionEncoder()
te_ary = te.fit(records).transform(records)
assert isinstance(te_ary, np.ndarray)  # 確認轉換後的型態為array

df_trans = pd.DataFrame(te_ary, columns=te.columns_)  # array轉DataFrame

df_trans
```

```python
# 利用apirori找出頻繁集frequent_itemsets
frequent_itemsets = apriori(df_trans, min_support=0.01, use_colnames=True)

frequent_itemsets
```

### 計算關聯規則


```python
from mlxtend.frequent_patterns import association_rules
```

#### 以最小支持度 min_support=0.006 找出 frequent_itemsets


```python
rules1 = association_rules(frequent_itemsets, metric="support", min_threshold=0.006)

rules1
```

#### 以 lift>1.000001 找出 rules


```python
rules2 = association_rules(frequent_itemsets, metric="lift", min_threshold=1.000001)

rules2
```

#### 以 confidence>=0.1 找出前置項為內衣時一同出現的字詞。


```python
rules3 = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)
rules3 = rules3[rules3["antecedents"] == {"內衣"}]

rules3
```

### 搭配 Jieba 執行 TF-IDF


```python
import jieba

# 設定 jieba 字典
jieba.set_dictionary("dict/dict.txt.big")
```

```python
from sklearn.feature_extraction.text import CountVectorizer  # 文本特徵提取
from scipy.sparse import csr_matrix  # 稀疏矩陣

vectorizer = CountVectorizer(
    tokenizer=jieba.lcut, stop_words=list(stops), lowercase=False
)

x = vectorizer.fit_transform(list(ptt_q2_df["內容"]))
word = vectorizer.get_feature_names_out()

assert isinstance(x, csr_matrix)

x, word
```

```python
from sklearn.feature_extraction.text import TfidfTransformer

tfidf = TfidfTransformer()
tfidf.fit(x)

tfidf_matrix = tfidf.transform(x)
tfidf_matrix
```

```python
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=word)  # type: ignore
tfidf_df
```

### 品牌與關鍵字相關性分析


```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# 找出 "Triumph" 關鍵字在 tfidf_df 的欄位位置
if "Triumph" not in tfidf_df.columns:
    raise ValueError("Triumph 不在關鍵字列表中")

# 取得 "Triumph" 關鍵字的 tfidf 向量
triumph_vec = tfidf_df["Triumph"].to_numpy().reshape(-1, 1)
triumph_vec
```

```python
# 計算所有關鍵字與 "Triumph" 的 cosine similarity
similarities = cosine_similarity(tfidf_df.values.T, triumph_vec.T).flatten()
similarities
```

```python
# 取得前五名最相關的關鍵字（排除 "Triumph" 本身）
triumph_idx = np.where(tfidf_df.columns == "Triumph")[0][0]
similarities[triumph_idx] = -1  # 排除自己

top5_idx = similarities.argsort()[-5:][::-1]
top5_keywords = tfidf_df.columns[top5_idx]
top5_scores = similarities[top5_idx]

# 顯示結果
for kw, score in zip(top5_keywords, top5_scores):
    print(f"{kw}: {score:.6f}")
```
