## 折線圖

折線圖（line chart）或曲線圖（curve chart）是由許多的資料點用直線連接形成的統計圖表，若看多個資料點之間的連線，會是折線。折線圖是許多領域都會用到的基礎圖表，折線圖類似散布圖，不過折線圖以 $X$ 軸為基礎，將 $X$ 軸上相鄰的資料點之間用直線連接。

通常用於呈現單一資料的時間趨勢，或是單一資料的不同群體的時間趨勢，故經常使用在時間序列類型的資料上。

這裡沒有時間序列的圖表，故隨機產生之。

```python
time_series_example = pl.DataFrame(
    {
        "date": ["2020-08-01", "2020-08-02", "2020-08-03", "2020-08-04"],
        "value1": [1, 2, 3, 4],
        "value2": [5, 4, 3, 2],
    }
)

alt.Chart(time_series_example).mark_line().encode(
    alt.X("date:T", title="日期"),
    alt.Y("value1:Q", title="值1"),
).interactive() + alt.Chart(time_series_example).mark_line(color="darkred").encode(
    alt.X("date:T", title="日期"),
    alt.Y("value2:Q", title="值2"),
).interactive()
```

### linear & cos/sin

```python
x = np.linspace(0, 10, 100)

alt.Chart(pl.DataFrame({
    "index": x,
    "sin(x)": np.sin(x),
    "cos(x)": np.cos(x),
})).transform_fold(
    fold=['sin(x)', 'cos(x)'],
).encode(
    alt.X("index:Q", title="x"),
    alt.Y("value:Q", title="y"),
    color="key:N",
    strokeDash="key:N",
).mark_line().interactive()
```

## 柱狀圖 (Bar Charts)

- 是一種以長方形或長方體的高度為變數的表達圖表，由一系列高度不等的縱向條紋表示資料分佈的情況，用來比較兩個或以上的價值（不同時間或者不同條件），只有一個變數，通常利用於較小的資料集分析。
- 柱狀圖可以用來比較資料之間的多少，可以用來觀察某一事件的變化趨勢，柱狀圖亦可橫向排列，或用多維方式表達。

```python
alt.Chart(df).encode(
    alt.X("chas:N", title=description_map["CHAS"]),
    alt.Y("tax", title=description_map["TAX"]),
    color="chas:N"
).mark_bar()
```

```python
alt.Chart(df).encode(
    alt.X("tax", title=description_map["TAX"]),
    alt.Y("chas:N", title=description_map["CHAS"]),
).mark_bar()
```

## 直方圖 (Histogram)

直方圖（Histogram）是一種對數據分布情況的圖形表示，是一種二維統計圖表，它的兩個坐標分別是統計樣本和該樣本對應的某個屬性的度量，以長條圖（bar）的形式具體表現。因為直方圖的長度及寬度很適合用來表現數量上的變化，所以較容易解讀差異小的數值。

```python
alt.Chart(df).mark_bar().encode(
    alt.X("tax", title=description_map["TAX"], bin=alt.Bin(maxbins=20)),
    alt.Y("count()", title="數量"),
)
```

## 圓餅圖 (Pie Chart)

```python
(
    alt.Chart(df)
    .mark_arc()
    .encode(
        alt.Tooltip("count()"),
        alt.Theta(
            "count(chas)",
            title="數量",
            stack=True,
        ),
        alt.Color("chas:N", title=description_map["CHAS"]),
    )
)
```

```python
(
    alt.Chart(df)
    .mark_arc()
    .encode(
        alt.Tooltip("count()"),
        alt.Theta(
            "count()",
            title="數量",
            stack=True,
        ),
        alt.Color("rad:N", title=description_map["RAD"]),
    )
)
```

## 箱型圖 (Box Plot)

```python
(
    alt.Chart(df)
        .mark_boxplot()
        .encode(
            alt.X("rad:N", title=description_map["RAD"]),
            alt.Y("tax:Q", title=description_map["TAX"]),
            alt.Color("chas:N", title=description_map["CHAS"]),
        )
        .properties(width=300)
)
```

## 子圖 (Subplot)

```python
base = alt.Chart(df).encode(
    alt.Theta("count()").stack(True),
    alt.Color("chas:N"),
)

(
    alt.Chart(df).mark_arc().encode(theta="count()", color="rad:N")
        | (
            base.mark_arc() + base.mark_text(radius=160, size=12).encode(text="count()")
        )
).interactive()
```
