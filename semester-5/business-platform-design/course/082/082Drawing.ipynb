```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['PingFang HK']
plt.rcParams['figure.dpi'] = 254
```

```python
listgrade = [
    [90, 99, 100],
    [99, 87, 94],
    [92, 85, 76],
    [87, 67, 72],
    [85, 89, 67],
]

gradedf = pd.DataFrame(listgrade,
                       columns=['數學', '英文', '物理'],
                       index=['No1', 'No2', 'No3', 'No4', 'No5'])

gradedf
```

```python
gradedf.plot()
plt.show()
```

將上面兩張圖改為折線圖在左上，長條圖在右下

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, 2, dpi=200)

gradedf.plot(kind='line', ax=ax[0,0])
gradedf.plot(kind='barh', ax=ax[1,1])

plt.show()
```
