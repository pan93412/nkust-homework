```python
import pandas as pd
```

# Load Data

```python
df = pd.read_excel('Online_Retail001.xlsx')

df
```

# Data Preparation

```python
df.describe()
```

```python
df["CustomerID"].isna().sum()
```

```python
df = df.loc[df['Quantity'] > 0]
df = df.dropna(subset="CustomerID")

df.describe()
```

```python
df["CustomerID"].isna().sum()
```

# Customer-Item Matrix

```python
customer_item_matrix = df.pivot_table(
    index='CustomerID', columns='StockCode', values='Quantity', aggfunc='sum'
)

customer_item_matrix = customer_item_matrix.map(lambda x: 1 if x > 0 else 0)
customer_item_matrix
```

# Collaborative Filtering

```python
from sklearn.metrics.pairwise import cosine_similarity
```

## User-based Collaborative Filtering

### User-to-User Similarity Matrix

```python
user_user_sim_matrix = pd.DataFrame(
    cosine_similarity(customer_item_matrix),
    columns=customer_item_matrix.index,
    index=customer_item_matrix.index
)

user_user_sim_matrix
```

### Making Recommendations

找出與 `12347.0` 這個顧客相似的顧客

```python
t = user_user_sim_matrix.loc[12347.0]
assert isinstance(t, pd.Series)
t= t.sort_values(ascending=False)

t.to_frame()
```

接著要找出 12347 曾經購買過的商品。

這裡要使用 `nonzero` 函數，它會回傳元素中不為零的整數索引值。

介紹：

```python
s = customer_item_matrix.loc[12347.0]
assert(isinstance(s, pd.Series))

s.to_frame()
```

```python
import numpy as np


s, = customer_item_matrix.loc[12347.0].to_numpy().nonzero() # 回傳 index
assert(isinstance(s, np.ndarray))

s
```

傳回 `index` 值，經由 `index` 值去找整筆資料⋯⋯

```python
customer_item_matrix.loc[12347.0].iloc[s] # 由 Series 的 index 值去找出是哪些值
```

`set` 為集合，找出 `12347` 購買過的產品

```python
items_bought_by_A = set(customer_item_matrix.loc[12347.0].iloc[s].index)

items_bought_by_A
```

找出與 12347 最類似的顧客購買過的產品

```python
s, = customer_item_matrix.loc[14790.0].to_numpy().nonzero()

items_bought_by_B = set(customer_item_matrix.loc[14790.0].iloc[s].index)

items_bought_by_B
```

找出兩者集合的差集

```python
items_to_recommend_to_B = items_bought_by_A - items_bought_by_B

items_to_recommend_to_B
```

從總資料找出 `items_to_recommend_to_B` 的產品編號與說明

```python
df.loc[
    df['StockCode'].isin(items_to_recommend_to_B),
    ['StockCode', 'Description']
]
```

去除重覆的產品編號，以及設定 index

```python
df.loc[
    df['StockCode'].isin(items_to_recommend_to_B),
    ['StockCode', 'Description']
].drop_duplicates().set_index('StockCode')
```

## 問題：Item-based Collaborative Filtering（產品導向協同過濾）

```python
item_item_matrix = df.pivot_table(
    index='StockCode', columns='CustomerID', values='Quantity', aggfunc='sum'
)

item_item_matrix = item_item_matrix.map(lambda x: 1 if x > 0 else 0)
item_item_matrix
```

### Item-to-Item Similarity Matrix

```python
item_item_sim_matrix = pd.DataFrame(
    cosine_similarity(item_item_matrix),
    columns=item_item_matrix.index,
    index=item_item_matrix.index
)

item_item_sim_matrix
```

### Making Recommendations

找出與 10120 這個產品類似的 10 個商品（包含 10120 這個商品共 10 個）

```python
t = item_item_sim_matrix.loc[10120]
assert isinstance(t, pd.Series)

t
```

```python
t1 = t.sort_values(ascending=False)
```

```python
top_10_similar_items = t1.iloc[:10].index
assert isinstance(top_10_similar_items, pd.Index)

list(top_10_similar_items)
```

# 找出產品清單

由彙總前資料找出前十名產品的產品編號與說明

```python
df.loc[
    df['StockCode'].isin(top_10_similar_items),
    ['StockCode', 'Description']
]
```

刪除重複的產品編號

```python
df.loc[
    df['StockCode'].isin(top_10_similar_items),
    ['StockCode', 'Description']
].drop_duplicates().set_index('StockCode')
```

按照前十名的順序排序

```python
df.loc[
    df['StockCode'].isin(top_10_similar_items),
    ['StockCode', 'Description']
].drop_duplicates().set_index('StockCode').loc[top_10_similar_items]
```
